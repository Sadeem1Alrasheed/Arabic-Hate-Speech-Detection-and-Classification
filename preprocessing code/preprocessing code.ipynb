```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from pyarabic.araby import strip_tashkeel, strip_tashkeel
```

```python
import re
import pyarabic.araby as araby
from nltk.tokenize import word_tokenize
```

```python
def preprocess_arabic_text(text):
    # Removal of Twitter shortcuts, such as usernames or retweets (RT)
    text = re.sub(r'RT\s|@\w+', '', text)

    # Deletion of special characters like @, %, and ~
    text = re.sub(r'[@%~]', '', text)

    # Removal of punctuation marks, emoticons, and numbers
    text = re.sub(r'[^\w\s]|[\d_]', '', text)

    # Deletion of URLs
    text = re.sub(r'http\S+', '', text)

    # Deletion of words containing only one character, such as "ص"
    text = ' '.join([word for word in text.split() if len(word) > 1])

    # Deletion of non-Arabic letters
    text = araby.strip_tashkeel(text)

    # Deletion of diacritics
    text = araby.strip_harakat(text)

    # Replace extra spaces, new lines, and underscores with a single space
    text = re.sub(r'\s+', ' ', text)

    # Convert "_" into space
    text = re.sub('_', ' ', text)

    return text
```

```python
# Apply the preprocessing function to the "comments" column in the dataframe
pd['Comments'] = pd['Comments'].astype(str).apply(preprocess_arabic_text)
```



> **3. Removal of Stop Words**



```python
from nltk.corpus import stopwords

# Assuming you have a list of Arabic stop words
arabic_stop_words = set(stopwords.words('arabic'))

# Assuming you have a function called arabic_stop_words
def remove_arabic_stop_words(tokens):
    filtered_tokens = [token for token in tokens if token.lower() not in arabic_stop_words]
    return filtered_tokens

# Apply the remove_arabic_stop_words function to the "Comments" column in the dataframe
t['Comments'] = t['Comments'].apply(remove_arabic_stop_words)

```



> **4. Stemming**



```python
import pandas as pd
from nltk.stem.isri import ISRIStemmer


# Create an instance of the ISRI stemmer
stemmer = ISRIStemmer()

# Apply the stemmer to the entire Arabic column in the dataframe
t['Comments'] = t['Comments'].astype(str).apply(lambda x: stemmer.stem(x))

# Print the dataframe with the stemmed Arabic text
print(t)
```



> **5. Normalization**



```python
# Create a function to perform normalization
def normalize_arabic_text(text):
    text = text.replace('1', 'ا').replace('!', 'ا').replace('ء', 'ا').replace('ة', 'ه').replace('ؤ', 'و')
    text = ''.join([text[i] for i in range(len(text)-1) if text[i] != text[i+1]])  # Delete duplicate letters
    text = text.replace('-', '')  # Delete the symbol "-"
    return text


```

```python
# Apply the normalization function to the entire Arabic column in the dataframe
t['Comments'] = t['Comments'].apply(normalize_arabic_text)

# Print the dataframe with the normalized Arabic text
print(t)
```



> **6. Delete Duplicated rows**
